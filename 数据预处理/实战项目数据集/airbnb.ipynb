纽约市Airbnb数据集

airbnb_NYC_2019.csv

简介：该数据集包含了2019年纽约市的Airbnb上线的房间情况。Airbnb是一个旅行房屋租赁社区，用户可通过网站或手机APP发布、搜索度假房屋租赁信息并在线预定。
变量含义：
- id：房间id
- name：房间名称
- host_id：房东id
- host_name：房东姓名
- neighbourhood_group：地区
- neighbourhood：街区
- latitude：纬度坐标
- longitude：经度坐标
- room_type：房间类型
- price：价格（美元）
- minimum_nights：最少预定夜晚数
- number_of_reviews：评论数量
- last_review：最新浏览时间
- reviews_per_month：每月浏览次数
- calculated_host_listings_count：房东挂出房子的数量
- availability_365：可预定房源的天数

目的：评估哪些街区的房源销售情况较好。

### 2.2.2 评估缺失值

有缺失值的变量是`name`、`host_name`、`last_review`和`reviews_per_month`

```python
o_data[o_data["name"].isnull()]
```

`name`和`host_name`对于房源销售量的影响，不会是主要因素，因此即使缺失对之后的分析影响也不会有。

```python
o_data[o_data["last_review"].isnull()]
```

`last_review`变量有10052条缺失的。

从上面显示的数据可以看出，`last_review`为NaN时`reviews_per_month`也为NaN并且`number_of_reviews`为0。我们猜想：
- (1)凡是`last_review`为NaN那么`reviews_per_month`也为NaN。
- (2)凡是`last_review`为NaN那么`reviews_per_month`也为NaN并且`number_of_reviews`为0。

接下来验证这个猜想。

```python
(o_data["last_review"].isnull() != o_data["reviews_per_month"].isnull()).sum()
```

由上的出结论：凡是`last_review`为NaN那么`reviews_per_month`也为NaN。

```python
o_data[o_data["last_review"].isnull() & (o_data["number_of_reviews"]!=0)]
```

在`last_review`为缺失值的情况下`number_of_reviews`不为0的情况不存在，也就是说`last_review`为缺失值的情况下`number_of_reviews`一定为0。

猜想(2)也的证。

# 3 清理数据

依据前面的评估需要做出如下清理：
1. `last_review`缺失值部分删除
2. `id`和`host_id`需要有整数转换为字符串
3. `last_review`需要转换为datetime数据类型

```python
c_data = o_data.copy()
c_data
```

**1. `last_review`缺失值部分删除**

```python
c_data.dropna(subset=["last_review"])
c_data = c_data.dropna(subset=["last_review"])
c_data
```

**2. `id`和`host_id`需要有整数转换为字符串**

```python
c_data["id"].astype(str)
```

```python
c_data["id"] = c_data["id"].astype(str)
c_data["id"]
```

```python
c_data["host_id"] = c_data["host_id"].astype(str)
```

**3. `last_review`需要转换为datetime数据类型**

```python
pd.to_datetime(c_data["last_review"])
```

```python
c_data["last_review"] = pd.to_datetime(c_data["last_review"])
c_data["last_review"]
```
